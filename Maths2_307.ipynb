# Fourier Series

## Aims

* Write a function as a Fourier series expansion and use the given formulae to obtain the Fourier coefficients, including for 
    1. continuous functions where only one integral is needed and 
    2. discontinuous (piecewise) functions that need breaking into a sum of integrals.
    3. the half-range series relating to physical problems


## Functions using Fourier Series


Any function with a fixed interval can be made from a series using sine and cosine *basis functions* using *Fourier Series* expansion.

For example a square wave can be made by adding different components together.

The figure below shows how to make up a *square-wave* function using the series: $y=\sin(x) + \dfrac{1}{3}\sin(3x) + \dfrac{1}{5}\sin(5x) + \dfrac{1}{7}\sin(7x)+\cdots$

<img width="800" src='Figures/fouriersum.png'>

**The red line shows the effect of adding the latest term to the previous result.**

This will prove useful when solving differential equations, which have solutions that are in terms of sinusoidal or (complex) exponential functions, but may have unusual initial conditions.

### Interpretation

The coefficients of the Fourier series show the strengths of the
frequency components of a periodic signal. A nice example is with sounds
or music.

Taking $x$ to be a fundamental frequency $f_0=1Hz$ in the
example above, the component signals would have frequencies
$1Hz, 3Hz, 5Hz, 7Hz\cdots$.

## Functions of Wavelength $2L$ (or period $T=2L$)

If a function $g(x)$ is defined over the range $-L \lt x \lt L$, the expression for the Fourier series is given by:

$$\eq{
g(x) = \frac{1}{2}a_0 &+ \sum\limits_{n=1}^{\infty} a_n \cos\left(\dfrac{n\pi x}{L}\right)\nonumber\\
& + \sum\limits_{n=1}^{\infty} b_n \sin\left(\dfrac{n\pi x}{L}\right),
}$$
 
With the coefficients given by the following integrals, with $1\leq n \leq \infty$:

$$\begin{align}
a_0 &= \frac{1}{L} \int\limits_{-L}^{L} g(x) {\mathrm{d}}x;\\
a_n &= \frac{1}{L} \int\limits_{-L}^{L} g(x) \cos\left(\dfrac{n\pi x}{L}\right) {\mathrm{d}}x, \quad\text{for}\quad n\geq1;\\
b_n &= \frac{1}{L} \int\limits_{-L}^{L} g(x) \sin\left(\dfrac{n\pi x}{L}\right) {\mathrm{d}}x, \quad\text{for}\quad n\geq1.\end{align}$$





---
Example: $g(\theta)=\theta$ from $-\pi<\theta<\pi$, so $2L=2\pi$ (or $L=\pi$).


Define the function to be $g(\theta)=\theta$ in the range $-\pi<\theta<\pi$ and imagine it repeats so that $g(\theta+2\pi)=g(\theta)$ (not considering outside this range).


### First  
find the coefficients:

$$\eq{
a_0 &= \frac{1}{{\pi}} \int\limits_{-\pi}^{\pi} \theta \, {\mathrm{d}}\theta;\\
&= \frac{1}{{\pi}}\left[ \frac{\theta^2}{2}\right]_{\theta=-\pi}^{\theta=\pi}\\
&=\frac{1}{2{\pi}}\left( \left( \pi^2 \right) -  (-\pi)^2 \right)\\
&=0\\
a_n = \frac{1}{L} \int\limits_{-L}^{L} g(\theta) \cos\left(\dfrac{n\pi \theta}{L}\right) {\mathrm{d}}x 
& = \frac{1}{{\pi}} \int\limits_{-\pi}^{\pi} {\theta} \cos({n\theta})\, {\mathrm{d}}\theta;\\
\\
\text{integrating by parts:}\quad &= 
\frac{1}{{\pi}}\left[\frac{\theta}{n} \sin{\left (n \theta \right )} + \frac{1}{n^{2}} \cos{\left (n \theta \right )}\right]_{-\pi}^{\pi}\\
&= 0\\
\\
b_n &= \frac{1}{{\pi}} \int\limits_{-\pi}^{\pi} {\theta} \sin({n\theta})\, {\mathrm{d}}\theta,\\
&=\frac{1}{{\pi}}\left[- \frac{\theta}{n} \cos{\left (n \theta \right )} + \frac{1}{n^{2}} \sin{\left (n \theta \right )}\right]_{-\pi}^{\pi}\\
&=\frac{1}{{\pi}}\left(- \frac{2 \pi}{n} \cos{\left (\pi n \right )} + \frac{2}{n^{2}} \sin{\left (\pi n \right )}\right)\\
&=\begin{cases}
\dfrac{2}{n} & \mbox{if: } n=1,3,5,\dots\\
- \dfrac{2}{n} & \mbox{if: } n=2,4,6,\dots 
\end{cases}
}$$

### Next
put these into the series expansion:

$$\eq{
g(\theta) = \frac{1}{2}a_0 &+ a_1 \cos(\theta) + a_2 \cos(2\theta) + a_3 \cos (3\theta) + \cdots\\
 &+ b_1 \sin(\theta) + b_2 \sin(2\theta) + b_3 \sin (3\theta) + \cdots,\\
 \\
= 0 &+  0 \cos(\theta) + 0 \cos(2\theta) + 0 \cos (3\theta) + \cdots\\
&+ \dfrac{2}{1} \sin(\theta) - \dfrac{2}{2} \sin(2\theta) + \dfrac{2}{3} \sin (3\theta) + \cdots\\
\\
\mbox{so: } g(\theta) = &\phantom{+}{2} \sin(\theta) -  \sin(\theta) + \dfrac{2}{3} \sin (3\theta) + \cdots
}$$

Odd and Even Functions
----------------------

The power-series for the sine function is:

$$\sin(\phi)= \phi - \dfrac{\phi^3}{3!} + \dfrac{\phi^5}{5!} + \ldots$$

* The sine function is made of only odd power terms


The power-series for the cosine function is:

$$
\cos(\phi)= 1 - \dfrac{\phi^2}{2} + \dfrac{\phi^4}{4!} + \ldots
$$

| | |
|-|-|

* The cosine function is made of only even power terms



-   An **odd** function has the rotational symmetry about
    the origin $-f(x) = f(-x)$.
-   An **even** function is one that has the mirror
    symmetry $f(x)=f(-x)$ about the y axis.
 
- **Odd** functions are made from only **sine** terms and **even** functions from only **cosine** terms in a Fourier series.
-   Some functions are neither odd nor even, and composed of a mixture
    of terms
    
| | |
|-|-|
|**Mix of sine and cosine terms!**![](Figures/nonefig.png) $f(a)\neq f(-a)$|$\mathbf{y=e^x}$ ![](Figures/expfig.png) $f(a)\neq-f(-a)$|
| | |

-   Multiplying: 
    * $f_{odd}\times f_{odd} \Rightarrow f_{even}$,
    * $f_{even}\times f_{even} \Rightarrow f_{even}$,
    * $f_{odd}\times f_{even} \Rightarrow f_{odd}$.
    

-   Integrating an *odd* function from $-a$ to $a$ gives
    the result of zero, because all contributions from the left cancel
    with those from the right of the $y$-axis.

![](Figures/oddint.png)

$$\begin{aligned}
    \int\limits_{-a}^{a} f_\text{odd} (x) {\mathrm{d}}x &=  \int\limits_{-a}^{0} f_\text{odd} (x) {\mathrm{d}}x +   \int\limits_{0}^{a} f_\text{odd} (x) {\mathrm{d}}x
     =  \int\limits_{-a}^{0} f_\text{odd} (x) {\mathrm{d}}x -  \int\limits_{0}^{a} f_\text{odd} (-x) {\mathrm{d}}x
    = 0\end{aligned}$$

-   The integral of an <span>*even*</span> function from $-a$ to $a$ is
    twice the integral from $0$ to $a$, since it is the same on the left
    and the right of the $y$-axis.

![](Figures/evenint.png)

$$\begin{aligned}
\int\limits_{-a}^{a} f_\text{even} (x) {\mathrm{d}}x &=  \int\limits_{-a}^{0} f_\text{even} (x) {\mathrm{d}}x +   \int\limits_{0}^{a} f_\text{even} (x) {\mathrm{d}}x
=  \int\limits_{-a}^{0} f_\text{even} (-x) {\mathrm{d}}x +   \int\limits_{0}^{a} f_\text{even} (x) {\mathrm{d}}x
= 2\int\limits_{0}^{a} f_\text{even} (x) {\mathrm{d}}x\end{aligned}$$

## Discontinous Functions:

For discontinuous (piecewise) functions, break the function into its
different sections and integrate them separately before summing the results.

### Example: Square Wave

A signal pulses on and off every period $T=2s$ (or $L=1s$). 
Express this as a Fourier series.

This is a discontinuous function that repeats every $2s$, so it needs to be defined in two parts over the interval $-1s < t < 1s$:

$$
\begin{align}
g(t) = \left\{ 
\begin{array}{l l}
0 & \quad\text{for}\quad -1s < t < 0,\\
1 & \quad\text{for}\quad  0 \leq t \leq 1s.
\end{array} 
\right.
\end{align}
$$

<img width=500 align='left' src='Figures/sqwav.png'>
<p style="clear: both;">

It can be seen from the graph that this will be based on a sine function shifted up by a constant so it sits on the x-axis: $g(t)=0.5+\sin\left(\frac{2\pi}{T}t\right)+\cdots$.

Since $L=1s$ the Fourier series can be written:
    
$$\begin{align*}
g(t) = \frac{1}{2}a_0 &+ a_1 \cos(\pi t) + a_2 \cos(2\pi t) + a_3 \cos (3\pi t) + \cdots\\
&+ b_1 \sin(\pi t) + b_2 \sin(2\pi t) + b_3 \sin (3\pi t) + \cdots.
\end{align*}$$

To obtain $a_0$:  

$\begin{align*}
a_0 &=  \frac{1}{L} \int\limits_{{-L}}^{{L}} g(x) \dd t\\ 
    &= 1 \left( \int\limits_{\bmaths{-1}}^{\bmaths{0}} {0} \dd t + \int\limits_{\bmaths{0}}^{\bmaths{1}} {1} \dd t \right)\\
    & = {0 + \big[t\big]_0^1}\\
    & = 1.
\end{align*}$


For the other terms multiply through by the cosine and sine functions:

$$
g(t)\cos(n\pi t) = \left\{ 
\begin{array}{l l}
{0} & \quad -1s < t < 0,\\
{(1)\cos(n\pi t)} & \quad  0 \leq t \leq 1s
\end{array} 
\right.
$$

and:

$$
g(t)\sin(n\pi t) = \left\{ 
\begin{array}{l l}
{0} & \quad -1s < t < 0,\\
{(1)\sin(n\pi t)} & \quad  0 \leq t \leq 1s
\end{array} 
\right.
$$

For the cosine terms: 

$$
\begin{align*}
a_n &= \frac{1}{L}\int\limits_{-L}^{L} g(t)\cos(n\pi t) \dd t\\
&= 1\left( {\int\limits_{-1}^0 0 \dd t} + {\int\limits_0^1 \cos(n\pi t) \dd t} \right)\\
%
&= \left[ \frac{\sin(n\pi t)}{n\pi}\right]_0^1 = 0  \quad \text{for} \quad n>0, 
\end{align*}
$$
    
since $\sin(0) = \sin(\pi)=\sin(2\pi)=\sin(3\pi)=\cdots=0$.

Similarly, for the sine terms:

$$
\begin{align*}
b_n &= \frac{1}{L}\int\limits_{-L}^{L} g(t)\sin(n\pi t) \dd x\\
&= 1\left( {\int\limits_{-1}^0 0 \dd t} + {\int\limits_0^1 \sin(n\pi t) \dd t} \right)\\
%
&= \left[ {-\frac{\cos(n\pi t)}{n\pi}}\right]_{{0}}^{{1}}\\
%
&=  {-\frac{\cos(n\pi)}{n\pi}+\frac{\cos(0)}{n\pi}}
\end{align*}
$$

For $n=1,3,5,\cdots:\ \cos(n\pi)=-1$;  
for $n=2,4,6,\cdots:\ \cos(n\pi)=1$;  
and $\cos(0)=1$ so:
    
$$
b_{1,3,5,\cdots} = \frac{(+1+1)}{n\pi}=\frac{2}{n\pi} \quad 
\text{and} \quad b_{2,4,6,\cdots} = \frac{(-1+1)}{n\pi} = 0
$$

Therefore, putting these back into the expansion of $g(t)$: 

$$\displaystyle g(t) = \frac{1}{2} + \frac{2}{\pi} \left( \frac{\sin(\pi t)}{1} + \frac{\sin(3\pi t)}{3} + \frac{\sin(5\pi t)}{5} + \frac{\sin(7\pi t)}{7} + \cdots \right)$$

## Half-range series for functions of length L

To obtain series representations for functions that are solutions to PDEs it is often useful to define the so-called *half-range* Fourier series.

This is used when the initial condition (at $t=0$) along the length is some distribution $u(x,0)=f(x)$ in the range $0\leq x\leq L$, with a boundary condition of $u(0,t)=u(L,t)=0$ at each end.

![](Figures/halfwave.png)

We imagine an extended function $g(x)=f(x)$ over $0\leq x\leq L$ and $g(x)=-f(x)$ in the range $-L<x<0$, making it an *"odd function"* (See Appendix).

The half-series in this case uses only the sine terms of the Fourier expansion: 

$$\eq{
g(x) = \sum\limits_{n=1}^{\infty} b_n \sin\left(\dfrac{n\pi x}{L}\right),
}$$

with the $b_n$ coefficients given by:
$$\begin{aligned}
b_n &= \frac{2}{L} \int\limits_{0}^{L} g(x) \sin\left(\dfrac{n\pi}{L}x\right) {\mathrm{d}}x, \quad\text{for}\quad n\geq1.
\end{aligned}$$

* This is due to the fact that $g(x)$ and $\sin(x)$ are both *odd functions*, so $g(x)\times\sin(x)$ is *even*, and therefore $\int_{-L}^{L}\bullet = 2\int_0^L\bullet$.
