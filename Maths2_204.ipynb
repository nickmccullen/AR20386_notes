# Solving First Order Ordinary Differential Equations

## Aims

1.  Separate and directly integrate simple 1st order ODEs.
2.  Verify solutions by differentiating and putting them back into the ODE.
3.  Use trial solutions and obtain the coefficients by differentiation and  substitution.
4. Obtain the Integrating Factor and use it to solve 1st order ODEs.

## Introduction 

First Order ODEs are used to model a number of engineering phenomena including flows of heat or other quantities. Flow is governed by systems attempting to reach equilibrium when there is a gradient in the field. Therefore first order terms appear in flow problems and solutions are exponential decays. 

We know that the exponential function $x=e^t$ has the property that differentiating it gives a function of the same form, so for exponential decay:

$$\eqnum{
x &= e^{-kt}\label{expdecay}\\
\Rightarrow\quad\DE{}{x}{t} &= -ke^{-kt}\\
\therefore\quad \DE{}{x}{t} &= -kx
}$$

So exponential functions can *solve* first order differential equations.

## Solution by Direct Integration


Differential equations (DEs) can sometimes be solved by directly
integrating them.

For example ${\dfrac{\mathrm{d}^{} y}{\mathrm{d}x^{}}}=x$ can
be directly integrated to obtain the solution $y=\frac{1}{2} x^2 + C$,
where $C$ is an arbitrary constant.

For a general DE of the form
$\ensuremath{\dfrac{\mathrm{d}^{} y}{\mathrm{d}x^{}}}=f(x)$ simply integrate the Right Hand Side (RHS):

$$\eq{
\DE{}{y}{x}&=f(x)\\
\\
y &= \int f(x) \dd x
}$$

## Separable differential equations


Differential equations can sometimes be separated out so that $x$ and
$y$ terms are on opposite sides of the equation. A differential equation
is described as *“separable”* if it can be rearranged into this form:

$$g(y)\ensuremath{\dfrac{\mathrm{d}^{} y}{\mathrm{d}x^{}}} = f(x)$$

It can then be directly integrated by multiplying both sides by
$\mathrm{d}x$, and integrating:

$$\eqnum{
\int g(y)\dd y = \int f(x) \dd x,\label{separable}
}$$

Ending up with a solution for $y$ as a function of $x$: $y = s(x)$

---
### Example: Given $\dot u = -k u$, find the solution $u(t)$


$$\eq{\DE{}{u}{t}&=-ku\\
\text{separate:}\qquad\frac{1}{u}\dd u&=-k\dd t\\
\text{integrate:}\qquad\int \frac{1}{u}\dd u&=\int-k\dd t\\
\ln (u) + c_1 &= -kt + c_2\\
}$$



$\eq{
\text{rearrange & collect constants:}\qquad\ln(u)&=-kt+C\\
\text{exponentiate:}\qquad u&=e^{-kt+C}\\
&=e^{-kt}e^C\\
e^C\text{ is just a constant, rename }A:\qquad u &= Ae^{-kt}
}$

---

## Trial Solutions


Many differential equations in engineering are of the form
$\ensuremath{\dfrac{\mathrm{d}^{} y}{\mathrm{d}x^{}}}=by$ and have 
exponentially growing or decaying solutions like: $y=Ae^{bx},$ where $A$
is a constant that depends on the initial conditions. Hence $y=Ae^{bx}$
is a *general solution* rather than a *particular solution*.

When you see a differential equation of this form it is a good idea to
try an exponential solution and try to work out the constants.

This can be done by differentiating the solution and comparing the
coefficients to those in the differential equation.

---

### Example of the cooling equation

Recall Newton's law of cooling:

$$\eq{
\DE{}{\theta}{t} = -k(\theta-T_s).
}$$

This has a similar form to the decay equation.

* Make a substitution $u=\theta-T_s$, $\dd u = \dd \theta$, and try the solution $u=Ae^{-kt}$.


1: Substituting $u=\theta-T_s$,$\quad$ $\dd u = \dd \theta$: 
    
$$\DE{}{u}{t} = -ku,$$
    
- which is of the form we are familiar with,
- we know that $u=Ae^{-kt}$ is a solution.

2: Trial the solution $u=Ae^{-kt}$ and substitute $u=\theta-T_s$ back in:
    
$$\eq{
    u &= Ae^{-kt}\\
    {\theta-T_s} &= Ae^{-kt}\\
    \therefore\quad \theta &= {Ae^{-kt}+T_s}
    }$$

- $T_s=5^\circ C$ is a known constant in this case, so:
    
$$\eq{ \theta &= {Ae^{-kt}+5} }$$

3: Now to find $A$ put in the *initial condition* $\theta(0)=30^\circ C$ at $t=0$:
    
$$\eq{
    \theta(0) &= {Ae^{-k(0)}+5}\\
    {30} &= {A + 5}\\
    \therefore\quad A &= {25} 
    }$$

Finally ending up with the particular solution to the differential equation at this initial condition.

$$\eq{
\theta(t) = {25e^{-kt}+5}
}$$

## Verifying Solutions to Differential Equations


To see that a function is a solution to a differential equation,
differentiate the solution and put it into both sides of the differential equation to see if it fits.

---

### Example: Show that the solution $\theta(t) = 25e^{-kt}+5$ really does satisfy the differential equation $\DE{}{\theta}{t} = -k(\theta-5)$.

Put the solution into both sides of the DE and see if it equates.

**LHS**

$$\eq{
{\DE{}{\theta}{t}} &= {\DE{}{}{t}\left({25e^{-kt}+5}\right)}\\
&= \bmaths{-25ke^{-kt}}
}$$

**RHS**

$$\eq{
{-k(\theta-5)} &= {-k\left({25e^{-kt}+5}-5\right)}\\
&= \bmaths{-25ke^{-kt}}
}$$

### Exercise of a falling body with viscous drag.

To find a solution to the equation:

$$\eq{\DE{}{v}{t} = 9.8 - \frac{v}{5}}$$

1. Find the general solution by separating and integrating it.
2. Use the initial condition $v_0=0$ at $t_0=0$ to find the constant of integration and turn the general solution into a particular solution.

Substitute $u=9.8-\dfrac{v}{5}$:

$${\DE{}{v}{t} = u}$$

$\DE{}{u}{v}={-\dfrac{1}{5}},\quad$  $\dd v = {-5 \dd u},\quad$now substitute this and

separate: $\qquad{\dfrac{1}{u}}\dd u = {-\frac{1}{5}}\dd t$


Integrate:

$${\int{\frac{1}{u}} \dd u = \int {-\dfrac{1}{5}} \dd t}$$

$${{\ln(u)} = {-\frac{t}{5}}}+c$$

Take $e^\bullet$ of both sides (remember $e^{\ln x} = x$):

$$\eq{
u&={e^{-t/5 + c}} 
}$$

Now rearrange this into the form $u=Ae^{\bullet}$ ($a^{b+d}=a^{b}a^{d}$ and $A=e^c$):

$$\eq{
u&={Ae^{-t/5}} 
}$$

Substitute $u=9.8-\dfrac{v}{5}$ back in:

$$\eq{
{9.8-\dfrac{v}{5}=Ae^{-t/5}}}$$

Should be able to rearrange this in terms of $v$:

$$\eq{
v(t) &= {5\times 9.8 - 5 A e^{-t/5}}\\
 &= 49 - B e^{-t/5}.
}$$

Setting $B=5A$ for convenience.

This is the **general solution**, where $B$ is a constant that depends on the initial conditions.

### Determining a Particular Solution

Put in the initial condition $v_0=0$ at $t_0=0$:

$$\eq{
v(0) = {49 - Be^{-0/5}}\\
{0} = {49 - B\times 1}\\
B = {49} 
}$$

So the *particular* solution for this initial condition is:

$$\eq{
v(t) &= {49 - 49e^{-t/5}}\\
 &= 49(1-e^{-t/5})
}$$


---

Once a possible solution has been identified, it should be checked that it *fits* ("solves") the differential equation.

---

#### Example: Verify that $v=49(1-e^{-t/5})$ satisfies the differential equation by putting it into the left hand side (LHS) and right hand side (RHS) and seeing if they are the same.

The differential equation is: $\DE{}{v}{t} = 9.8-\dfrac{v}{5}$, so putting the solution into both sides:

**LHS**:

$$\eq{
v &= 49(1-e^{-t/5})\\
\DE{}{v}{t} &= {\frac{49}{5}e^{-t/5}}\\
&= {9.8 e^{-t/5}}
}$$

**RHS**:

$$\eq{
9.8-\dfrac{v}{5} &= 9.8-\frac{1}{5}{49(1-e^{-t/5})}\\
&= 9.8-{9.8(1-e^{-t/5})}\\
&= 9.8-{9.8+9.8e^{-t/5}}\\
&= {9.8 e^{-t/5}}
}$$

So both sides equate and the solution satisfies the equation.

## Solving more complicated 1st Order ODEs

The *Standard Form* of 1st order ODEs is:

$$\eqnum{\DE{}{y}{x} + f(x)y = r(x).}$$

Next look at the case where $f(x)$ can be a function rather than a constant $k$.

## Homogeneous Differential Equations

*Homogeneous* ordinary differential equations appear in the form: 

$$\DE{}{y}{x} + f(x)y = 0,$$
where the right hand side is still simply $r(x)=0$.




Separating this we can find the general solution: 

$$\eq{
\DE{}{y}{x} &= -f(x)y,\\
\int \frac{1}{y} \dd y &= -\int f(x) \dd x,\\
\ln(y) &= -\int f(x) \dd x + c,\\
}$$

To simplify the
notation define a function $h(x)$ to be: 

$$\eqnum{
h(x)=\int f(x) \dd x
}$$

Now exponentiate the general
solution: 

$$\eq{
y &= e^{-h(x) + c},\\
 &= e^{-h(x)}e^{c},\\
&=Ae^{-h(x)},
}$$

by re-writing $e^c=A$.

This gives the ***general solution*** to homogeneous differential
equations: 

$$\eqnum{
{y=Ae^{-h(x)},}\label{gensol1}
}$$

where $\displaystyle h(x)=\int f(x) \mathrm{d}x$
is worked out case-by-case.

---

#### Example: $y'=-ky$

Re-writing this in the *standard form*:

$$\eq{
\DE{}{y}{x}+f(x)y&=0,\\
\DE{}{y}{x}+ky &= 0,
}$$

with $f(x)=k$ just a constant.
Now to find $h(x)$:

$$\eq{
h(x)&=\int f(x) \dd x\\
&=\int k \dd x,\\
&=kx.}$$

So the general solution is:

$$\eq{
y&=Ae^{-h(x)}\\
&=Ae^{-kx},
}$$

as before.

---

## Solving Inhomogeneous First Order ODEs

*Inhomogeneous* equations have a non-zero right hand term:

$\eqnum{
{y' + f(x)y = r(x)}\label{nh1}.
}$

## Integrating Factors 

The integrating factor method is useful for finding the solutions of inhomogeneous ODEs.

Define the ***integrating factor*** to be $I=e^{h(x)}$; where $h(x)=\int f(x)\dd x$, as before.



An expression for the ***general solution*** is found using the folllowing formula (see Appendix for derivation):

$$\eqnum{
y &= \frac{1}{I}\left[ \int r(x)I \dd x\right].\label{gs1}
}$$

-   You will **not** need to memorise this, just know how to use it.

---

### Step-by-step Example: $t\dot{y} + 2y = 4t^2$ with $y(1)=4$.

Rearrange into the standard form by dividing by $t$:

$$\eq{
\DE{}{y}{t} + \frac{2}{t}y &= 4t
}$$

Now it is in the form above, with $f(t)=\dfrac{2}{t}$ and $r(t)=4t$.

**First** find the integrating factor $I=e^{h(t)}$, by integrating $f(t)$:

$$\eq{
h(t) &= \int {\frac{2}{t}} \dd t\\
&= {2 \ln (t) + c}\\
\therefore\quad e^{h(t)} &= {e^{2\ln(t)+c}}\\
e^{h(t)} &= {Ae^{2\ln(t)}}\\
&= {A\left(e^{\ln(t)}\right)^2}\\
&= At^2
}$$

This is because $e^t$ and $\log(t)$ are reciprocal functions, therefore replace $(e^{\ln(t^2)}=)e^{2\ln(t)}=e^{\ln(t)+\ln(t)}=e^{\ln(t)}e^{ln(t)}.$

Since in the IF formula we both multiply and divide by $I$, any constant multiplying factors will be lost so we can ignore them and end up with: 

$$I=t^2.$$ 

We can see that $r(t)=4t$, so put these into the formula:

$$\eq{
y &= \frac{1}{I}\left[ \int r(t)I \dd t\right],\\
 &= {t^{-2}}\left[ \int {4t (t)^2} \dd t\right]\\
 &= {t^{-2}}\left[ { t^4} + c\right]\\
  &=  t^2 + \frac{c}{t^2},
}$$

as before.
Now use the initial condition $y(1)=4$ to find a particular solution:

$$\eq{
4 &= 1^2 + \frac{c}{1^2},\\
c &= 3\\
y &= t^2 + \frac{3}{t^2}.
}$$

## Dealing with Non-Integrable Functions

The integrating factor method relies on being able to integrate $\int f(x)$ and $\int r(x) I$.
However, sometimes this is not possible (consider integrating $\int e^{x^2}\dd x$).


*  A power series could be used to expand any function that is difficult to integrate  into a polynomial $-$ as we will see next $-$ then integrate the terms separately and sum back together. 


## Extra: Deriving Integrating Factors 

To understand how to derive the IF method, rearrange the general solution for the homogeneous ($r(x)=0$) case, by multiplying through by $e^{h(x)}$:

$$\eq{
y&=Ae^{-h(x)},\nn\\
ye^{h(x)}&=A
}$$

and differentiate both sides:

$$\eq{
{\DE{}{}{x}\big\{ye^{h(x)}\big\}}&=\DE{}{A}{x}.\\
&= 0,\nn
}$$

The product rule and the chain rule can be used (try this!) to expand the LHS:

$$\eq{
 \DE{}{}{x}\big\{ye^{h(x)}\big\}&= e^{h(x)}\DE{}{y}{x} + yf(x)e^{h(x)}.
}$$

Terms of the form $y'+yf(x)$ can therefore be multiplied by a factor $I=e^{h(x)}$ to give $y'I+yf(x)I$ and simplified to $\DE{}{}{x}\big\{yI\big\}$ using the above identity. 


## Extra: Full Method for Solving Inhomogeneous ODEs

The left hand side of first order linear differential equations of the form $y'+yf(x)=r(x)$ can therefore be simplified by multiplying through by $I=e^{h(x)}$ and making this substitution to the LHS:

$$\eq{y'I + yf(x)I = \DE{}{}{x}\big\{yI\big\}}$$



First determine the integral:

$$\eqnum{
h(x)=\int f(x) \dd x,
}$$

then multiply through by the integrating factor $I=e^{h(x)}$ and substitute the LHS using the above substitution: 

$$\eqnum{
y'I + f(x)yI &= r(x)I,\nn\\
\DE{}{}{x}\{yI\} &= r(x)I.\label{IF}
}$$

The general solution is then given by integrating both sides, then dividing by $I$:

$$\eq{ \int \DE{}{}{x}\{yI\} \dd x &= \int r(x)I\ \dd x,\\
yI &= \int r(x)I\ \dd x,\\
y &= \frac{1}{I}\left[ \int r(x)I\ \dd x\right].
}$$
