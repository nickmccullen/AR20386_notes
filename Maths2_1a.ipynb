# Power Series and Complex Numbers

## Approximating using Polynomial Series

### Aims:
1. Understand that "polynomial" or "power series" can be used to approximate a function at a point to any degree of precision.
2. Be able to derive the coefficients of the series for any repeatably differentiable function.
3. Appreciate some applications of these series in solving otherwise difficult or impossible problems.

### Infinite Series

Some infinite series converge to a numerical value, and are an approximation to that value when truncated to a finite number of terms. For example the constant $\pi$ can be obtained using the infinite sequence:

$$\dfrac{\pi}{4} =  1 - \dfrac{1}{3} + \dfrac{1}{5} - \dfrac{1}{7} + \dfrac{1}{9} - \cdots$$

This can be written using summation notation: 

$$\pi = 4 \sum_{k=0}^{\infty}\dfrac{(-1)^k}{2k+1}$$

Since $(-1)^k$ takes the values: $\quad(-1)^0=1,\quad (-1)^1=-1,\quad (-1)^2=1,\quad (-1)^3=-1,\quad\ldots$

### Power Series Expansion

It is also possible to represent functions as a series of powers with a particular set of coefficients $a_n$, such that:

$$\eq{
f(x) &= a_0 + a_1 x + a_2 x^2 + a_3 x^3 + a_4 x^4 + a_5 x^5 + \ldots\\
&= \sum_{n=0}^{\infty}a_n x^n
}$$

where the numerical value of the function $f(x)$ at a particular value of $x$ is approximated by the finite series and each term adds a bit more accuracy to the approximation.


The first coefficient can be obtained by setting $x=0$:

$\eq{
f(0) &= a_0 + a_1 (0)  + a_2 (0)^2 + a_3 (0)^3 + \ldots\\
f(0) &= a_0.
}$

So $a_0$ is simply the value of the function itself at $x=0$.

If the series is differentiated, the $a_1$ term no longer has an $x$ in it:

$\DE{}{f}{x} =  a_1 + 2 a_2 x + 3 a_3 x^2 + 4 a_4 x^3 + 5 a_5 x^4 + \ldots$

The other terms can again be removed by setting $x=0$

$f'(0) =  a_1 + 2 a_2 (0) + 3 a_3 (0)^2 + 4 a_4 (0)^3 + 5 a_5 (0)^4 + \ldots,$

so $a_1=f'(0)$.

The remaining coefficients $a_n$ are obtained by repeated differentiation of the function that is being approximated and then setting $x=0$.

For the $a_2$ and $a_3$ coefficients:

$\eq{
f''(x) &=   (2)a_2 + (3\times2) a_3 x + (4\times3) a_4 x^2 + (5\times4) a_5 x^3 + \ldots\\
f'''(x) &=   (3\times2) a_3  + (4\times3\times2) a_4 x + (5\times4\times3) a_5 x^2 + \ldots\\
}$
and so on.  

So the coefficients are:

$\eq{
a_2 &= \dfrac{f''(0)}{2}\\
a_3 &= \dfrac{f'''(0)}{3\times2}.
}$

This continues, so the general formula for the coefficients is given by:

$$a_n = \dfrac{1}{n!}f^n(0)\qquad\equiv\DE{n}{f(x)}{x},\, x=0$$

#### Example: expanding $f(\theta) = {e^{\theta}}$ to 6 terms (including $a_0$):

$a_0=e^{(0)}=1$

##### For $a_1$:

$\DE{}{e^\theta}{\theta} = e^\theta$

So applying the formula for coefficients:

$a_1 = \dfrac{1}{1!}f'(0) = 1\times e^0 = 1$

##### For $a_2$: 

$f''(\theta) = \DE{}{}{\theta}\left\{ \DE{}{e^\theta}{\theta} \right\} = e^\theta$

Applying the formula:

$a_2 = \dfrac{1}{2!}f''(0) = \dfrac{1}{2} e^0 = \dfrac{1}{2}$

##### For $a_3$: 

$\eq{
f'''(\theta) &= e^\theta\\
a_3 &= \dfrac{1}{3!}f''(0) = \dfrac{1}{6} e^0 = \dfrac{1}{6}
}$

##### For other coefficients: 

$f^{(n)}(\theta) = \DE{n}{e^\theta}{\theta} = e^\theta$

So:

$\eq{
a_4 &= \dfrac{1}{4!}f^{(4)}(0) = \dfrac{1}{24} e^0 = \dfrac{1}{24}\\
a_5 &= \dfrac{1}{5!}f^{(5)}(0) = \dfrac{1}{120} e^0 = \dfrac{1}{120}\\
}$

The function $e^\theta$ can therefore be represented as an infinite sequence:

$$\eq{
e^\theta &= 1 + {}\theta + \dfrac{1}{2}{}\theta^2 + \dfrac{1}{3!}{}\theta^3 + \dfrac{1}{4!}{}\theta^4 + \dfrac{1}{5!}{}\theta^5 + \ldots},$$
which can be truncated at the desired precision.

### Application: Integrating Non-Integrable Functions

Some integrals cannot be evaluated analytically. For example $e^{-x^2}$ has no integral that can be written down in a single finite expression. However, it can be approximated using a power series and this can be integrated to give a numerical result.

#### Example: integrating the standard normal distribution to obtain values for the cumulative distribution function

The probability density function for the normal distribution is given by: $\displaystyle\dfrac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$.

The cumulative distribution function (used to estimate probabilities) is the intagral of $f(x)$, but this can not be evaluated analytically. Instead numerical approximations are used, such as power series expansion.

Use the expansion for $e^\theta$ to four terms and integrate in order to obtain an approximation to the following integral to 3 decimal places:  
$$\displaystyle I = \int_0^1 e^{-z^2/2} ~\dd z.$$



$e^\theta \approx 1 + \theta + \dfrac{\theta^2}{2!} + \dfrac{\theta^3}{3!} +\ldots$


Set $\theta=-z^2/2$:

$\eq{
e^{-z^2/2} &\approx 1 - \dfrac{z^2}{2} + \dfrac{z^4}{2^2\times2!} - \dfrac{z^6}{2^3\times 3!} +\ldots\quad\\
 &= 1   - \frac{z^{2}}{2} + \frac{z^{4}}{8} - \frac{z^{6}}{48} + \ldots
}$

Integrating:

$\eq{
\int_0^1 e^{-z^2/2} \,\dd z &\approx \int_0^1 1 \,\dd z - \int_0^1 \dfrac{z^2}{2} \,\dd z + \int_0^1 \dfrac{z^4}{8} \,\dd z - \int_0^1 \dfrac{z^6}{48} \,\dd z + \ldots\\
&= \left[ x -  \dfrac{x^3}{6}  +  \dfrac{x^5}{40}  -  \dfrac{x^7}{336} + \ldots\right]_0^1  \\ 
&= 1 -  \dfrac{1}{6}  +  \dfrac{1}{40}  -  \dfrac{1}{336} + \ldots\\
&=0.855
}$

Hence, values can be obtained for use in probability estimation.

#### How many terms would be needed to obtain the approximation to 2 d.p.?


$0.855\approx0.86; \quad 1 -  \dfrac{1}{6}  +  \dfrac{1}{40}  =0.858\approx0.86; \quad 1 -  \dfrac{1}{6}   =0.833\approx0.83$

So the fourth term makes no difference and only three terms are needed.

### Imaginary Numbers


Define a new *imaginary number* $i$ such that $i^2=-1$, and call multiples of $i$ *“imaginary numbers”*. 

-   The inverse is $i \equiv \sqrt{-1}$, so replace $\sqrt{-1}$ with $i$
    wherever you see it.
    
-   Use it in place of a real number, but whenever we square it to get $i^2=-1$ we can go back to real numbers. E.g. if $B=i\sqrt{k}$ then $B^2=-k$.

All calculations involving ${i}$ use the same rules of algebra, calculus
etc. as usual, except that $i^2$ is replaced by $-1$. Calculations can
often be a lot simpler using $i$, as you will see.

#### Note:

Engineers often use $j$ to avoid confusion with electrical current
    ($I$), but these notes will stick with using $i$.

### Solving Quadratics

Quadratic equations of the form $\eq{az^2 + bz + c=0}$ can be solved using the formula:

$$
\eqnum{z = \dfrac{-b\pm \sqrt{b^2-4ac}}{2a}.}
$$

---

#### Example: 

Find the roots of $z^2 + 2z + 5 = 0$

$$\eq{z &= \dfrac{-(2)\pm\sqrt{{(2)^2 - 4(1)(5)}}}{2(1)}\\
&= \dfrac{-2\pm\sqrt{{-16}}}{2}\\
&= \dfrac{-2\pm\sqrt{{(16)(-1)}}}{2}\\
&= \dfrac{-2\pm\sqrt{{16}}\sqrt{{-1}}}{2}\\
&= \dfrac{-2\pm{4}\sqrt{{-1}}}{2}\\
&= -1\pm2\sqrt{-1} = -1\pm2i
}$$

---

### Complex Numbers


A **complex number** is a number with a real part and an
imaginary part.
Thus the *real* part of $z=7+3i$ is $7$ and the
*imaginary* part is $3$.

This is written:

$$\begin{array}{l c r}
Re ( 7 + 3i ) = 7\\
Im ( 7 + 3i ) = 3
\end{array}$$

### Complex Arithmetic


#### Addition and Subtraction

Simply add or subtract the term of the same type

---

#### Example: $a=3+2i, b=2+i$

{$a+b$}

$a + b = 3 + 2i + 2 + i = (3 + 2) + (2i + i) = 5 + 3i$

{$a-b$}

$a - b = (3 + 2i) - (2 + i) = (3 - 2) + (2i - i) = 1 + i$}

---

#### Multiplication

Multiply out as you would for an algebraic expression

---

#### Example: $(3 + 2i)(2 + i)$

$$\eq{
(3 + 2i) (2 + i) &= (3)(2) + (3)(i) + (2i)(2) + (2i)(i)\\
&= 6 + 7i + (-1)2\\
&= 4 + 7i
}$$

---

### **Conjugate Complex Numbers:**


The *conjugate* ${\overline{z}}$ of a complex number ${z}$
is the same except for a sign change on the imaginary part. That is:
${z = a {+} bi}$ and ${\overline{z} = a {-} bi}$ are conjugate.
Multiplying the two together gives a real answer:

$(a+bi)(a-bi) = {{\color{\excolor}\large a^2 - b^2i^2 + abi - abi\color{black}}} = a^2+b^2$

### Dividing by Complex Numbers

Need to convert the denominator into a real number: so multiply bottom
and top by the complex conjugate of the denominator.

$$\eq{
\dfrac{z_1}{z_2} &=\dfrac{z_1}{z_2}\dfrac{\bar{z_2}}{\bar{z_2}}\\
&=\dfrac{z_1}{(c+di)}\dfrac{\bar{z_2}}{(c-di)}\\
&=\dfrac{z_1\bar{z_2}}{c^2+d^2}
}$$

---

#### Example: $\dfrac{2-4i}{4+2i}$

$=\dfrac{(2-4i)}{(4+2i)}\dfrac{\bmaths{(4-2i)}}{\bmaths{(4-2i)}}$
$=\dfrac{{8-20i-8}}{{16+4}}$
$={-i}$

### Exponential Form of Complex Numbers: $z = r e^{i\phi}$

* To show where this comes from we need power series expansions...


The Cartesian form of complex numbers is $z=x+iy$.

The polar form of complex numbers is obtained using polar coordinates:

$$\eq{x &= r \cos(\phi)\\
y &= r \sin(\phi)\\
z &= x + iy = r(\cos(\phi) + i\sin(\phi))
}$$

#### Power Series expansion of $\cos (\phi)+ i\sin(\phi)$

The power series expansion of $\cos(\phi)$ is (exercise):

$$\eq{
\cos(\phi)&{= 1 - \dfrac{\phi^2}{2!} + \dfrac{\phi^4}{4!} - \ldots}
}$$

The power series expansion of $\sin(\phi)$ is (exercise):

$$\eq{
{\sin(\phi)}&{= \phi - \dfrac{\phi^3}{3!} + \dfrac{\phi^5}{5!} - \ldots}
}$$

Combining these two results as $\cos(\phi)+i\sin(\phi)$:

$$\eq{
\cos(\phi)+i\sin(\phi)&=
\left(1 - \dfrac{\phi^2}{2!} + \dfrac{\phi^4}{4!} + \ldots\right)
+i\left(\phi - \dfrac{\phi^3}{3!} + \dfrac{\phi^5}{5!} + \ldots\right)\nn
}$$

#### Power Series expansion of $e^{i\phi}$

Taking the expansion of $e^\theta$:

$$e^\theta = 1 + \theta + \dfrac{\theta^2}{2} + \dfrac{\theta^3}{3!} + \dfrac{\theta^4}{4!} + \dfrac{\theta^5}{5!} + \ldots$$

Set $\theta={i\phi}$:

$$
{e^{i\phi}} = { 1 + i\phi + \dfrac{(i\phi)^2}{2} + \dfrac{(i\phi)^3}{3!} + \dfrac{(i\phi)^4}{4!} + \dfrac{(i\phi)^5}{5!} + \ldots}
$$

Replace each $i^2=-1$.

Given that  
$i^3=(i^2)(i)=-i$,  
$i^4 = (i^2)(i^2) = (-1)(-1) = 1$,  
$\ldots$,  
it can be seen that $i^n$ follows the pattern $1,i,-1,-i,1,\dots$ for $n=0,1,2,3,4,\dots$

$$\eq{{e^{i\phi}}
&={ 1 + i\phi + \dfrac{(i\phi)^2}{2} + \dfrac{(i\phi)^3}{3!} + \dfrac{(i\phi)^4}{4!} + \dfrac{(i\phi)^5}{5!} + \ldots}\nn\\
&={ 1  + i\phi  - \dfrac{\phi^2}{2}  - i\dfrac{\phi^3}{3!}  + \dfrac{\phi^4}{4!} + i\dfrac{\phi^5}{5!}  + \ldots}
}$$

The real and imaginary parts now can be separated out:

$$\eq{
{e^{i\phi}}
&= \left(1 - \dfrac{\phi^2}{2!} + \dfrac{\phi^4}{4!} + \ldots\right)
+i\left(\phi - \dfrac{\phi^3}{3!} + \dfrac{\phi^5}{5!} + \ldots\right)\nn
}$$

This is the same as the expansion of $\cos(\phi) + i\sin(\phi)$ above, proving that $e^{i\phi}=\cos(\phi) + i\sin(\phi)$ the exponential form of a complex number is: 

$z = r e^{i\phi}$

### The complex conjugate:

$x - iy = \cos(\phi) - i \sin(\phi) = e^{-i\phi}$

### Multiplying Complex Exponentials

$z_1z_2 = r_1r_2e^{i\phi_1}e^{i\phi_2} = r_1r_2e^{i(\phi_1+\phi_2)}$: a rotation on the argand diagram.

$z\bar{z} = r^2e^{i\phi}e^{-i\phi} = r^2e^{i\phi-i\phi} = r^2[e^0=1] = a^2+b^2$

### Finally: Euler's Identity  $-$ The Most Amazing Equation!

What happens when $\theta=\pi$ is put into
$e^{i \theta} = \cos(\theta) + i \sin(\theta)$?

$e^{i\pi} = \cos(\pi) + i \sin(\pi),$

Think about what this means geometrically on an Argand diagram.  
However, 
$\cos(\pi) = -1$, and $\sin(\pi) = 0$, so

$e^{i\pi} = -1 + 0i$. 

Or to put it another way:

$e^{i \pi} + 1 = 0$

which links five of the most important numbers in
maths in one expression!

### Appendix: Very Useful Identities

The following can be used to convert between trigonometric expressions and complex exponentials, which can be used instead of memorising many trig identities...

$\cos(\phi)=\dfrac{e^{i\phi} + e^{-i\phi}}{2}$,

$\sin(\phi)=\dfrac{e^{i\phi} - e^{-i\phi}}{2i}$

---

#### Derivation:

$e^{i\phi}+e^{-i\phi}=[\cos(\phi)+i\sin(\phi)]+[\cos(\phi) - i \sin(\phi)]=2\cos(\phi).$

$e^{i\phi}-e^{-i\phi}=[\cos(\phi)+i\sin(\phi)]-[\cos(\phi) - i \sin(\phi)]=2i\sin(\phi).$
